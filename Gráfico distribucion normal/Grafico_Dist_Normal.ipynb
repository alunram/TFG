```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores x para el gráfico
x = np.linspace(-5, 5, 1000)

# Calcular los valores de la función de densidad de probabilidad (pdf) para una distribución normal estándar
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Crear el gráfico
plt.plot(x, y, label='Distribución normal estándar')
plt.title('Campana de Gauss de la Distribución Normal Estándar')
#plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.legend(loc='upper right', fontsize='small')  # Cambia la posición del texto del label
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

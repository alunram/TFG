```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores x para el grfico
x = np.linspace(-5, 5, 1000)

# Calcular los valores de la funcin de densidad de probabilidad (pdf) para una distribucin normal estndar
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Crear el grfico
plt.plot(x, y, label='Distribucin normal estndar')
plt.title('Campana de Gauss de la Distribucin Normal Estndar')
#plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.legend(loc='upper right', fontsize='small')  # Cambia la posicin del texto del label
plt.grid(True)

# Mostrar el grfico
plt.show()
```

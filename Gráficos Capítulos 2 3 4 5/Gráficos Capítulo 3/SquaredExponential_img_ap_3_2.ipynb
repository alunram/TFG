```python
import numpy as np
import matplotlib.pyplot as plt

# Calcula la función de covarianza entre x e y, con parámetro length_scale
def squared_exponential_covariance(xp, xq, length_scale):
    sqdist = np.sum(xp**2, 1).reshape(-1, 1) + np.sum(xq**2, 1) - 2 * np.dot(xp, xq.T)
    return np.exp(-0.5 / length_scale**2 * sqdist)

# Define el espacio: [-5,5] x [-5,5]
x = np.linspace(-5, 5, 100).reshape(-1, 1)

# Calcula la matriz de covarianza para length_scale = 1
length_scale_1 = 1
K1 = squared_exponential_covariance(x, x, length_scale_1)

# Calcula la matriz de covarianza para length_scale = 2
length_scale_2 = 2
K2 = squared_exponential_covariance(x, x, length_scale_2)

# Dibuja los gráficos
plt.figure(figsize=(16, 6))

# Gráfico para length_scale = 1
plt.subplot(1, 2, 1)
plt.imshow(K1, extent=(-5, 5, -5, 5), origin='lower', cmap='viridis')
cb1 = plt.colorbar(label='Covarianza')
cb1.ax.yaxis.label.set_fontsize(14)

plt.title('Squared Exponential (length_scale = 1)', fontsize=16)
plt.xlabel(r'$x_p$', fontsize=18)  # LaTeX syntax
plt.ylabel(r'$x_q$', fontsize=18)  # LaTeX syntax

# Gráfico para length_scale = 2
plt.subplot(1, 2, 2)
plt.imshow(K2, extent=(-5, 5, -5, 5), origin='lower', cmap='viridis')
cb2 = plt.colorbar(label='Covarianza')
cb2.ax.yaxis.label.set_fontsize(14)

plt.title('Squared Exponential (length_scale = 2)', fontsize=16)
plt.xlabel(r'$x_p$', fontsize=18)  # LaTeX syntax
plt.ylabel(r'$x_q$', fontsize=18)  # LaTeX syntax

plt.tight_layout()
plt.show()
```

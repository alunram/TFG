Primer gráfico: Función logística

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Crear valores de x
x = np.linspace(-10, 10, 100)

# Calcular los valores de la función sigmoide para los valores de x
y = sigmoid(x)

# Dibujar la función sigmoide
plt.plot(x, y, label='Función Logística', color='blue')
plt.title('Función Sigmoide utilizada en Regresión Logística')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Segundo gráfico: función probit

```python
from scipy.integrate import quad

# Define la función de densidad de probabilidad de una distribución normal estándar.
def standard_normal_pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)

# Utiliza quad de SciPy para calcular numéricamente la integral desde menos infinito hasta zz de la función de densidad de probabilidad
def probit(z):
    result, _ = quad(standard_normal_pdf, -np.inf, z)
    return result

# Crear valores de x
x = np.linspace(-5, 5, 100)

# Calcular los valores de la función probit para los valores de x
y = [probit(xi) for xi in x]

# Dibujar la función probit
import matplotlib.pyplot as plt
plt.plot(x, y, label='Función Probit', color='purple')
plt.title('Función Sigmoide utilizada en Regresión Probit')
plt.xlabel('z')
plt.ylabel('Probit(z)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def standard_normal_pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)

def probit(z):
    result, _ = quad(standard_normal_pdf, -np.inf, z)
    return result

# Crear valores de x para ambas funciones
x = np.linspace(-5, 5, 100)

# Calcular los valores de la función sigmoide para los valores de x
y_sigmoid = sigmoid(x)

# Calcular los valores de la función probit para los valores de x
y_probit = [probit(xi) for xi in x]

# Dibujar ambas funciones en el mismo gráfico
plt.plot(x, y_sigmoid, label='Función Logística', color='blue')
plt.plot(x, y_probit, label='Función Probit', color='purple')

plt.title('Comparación de Funciones Sigmoide', fontsize = 16)
plt.xlabel('z', fontsize = 15)
plt.ylabel('Valor de la Función', fontsize = 15)
plt.legend(fontsize = 13)
plt.grid(True)
plt.show()
```

Primer gráfico: función de densidad de la normal

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores x para el gráfico
x = np.linspace(-5, 5, 1000)

# Calcular los valores de la función de densidad de probabilidad (pdf) para una distribución normal estándar
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Crear el gráfico
plt.plot(x, y, label='Distribución normal estándar')
plt.title('Campana de Gauss de la Distribución Normal Estándar')
#plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad', fontsize=13)
#plt.legend(loc='upper right', fontsize=13)  # Cambia la posición del texto del label
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Segundo gráfico: Función logística

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Crear valores de x
x = np.linspace(-10, 10, 100)

# Calcular los valores de la función sigmoide para los valores de x
y = sigmoid(x)

# Dibujar la función sigmoide
plt.plot(x, y, label='Función Logística', color='blue')
plt.title('Función Sigmoide utilizada en Regresión Logística', fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

Tercer gráfico: función probit

```python
from scipy.integrate import quad

# Define la función de densidad de probabilidad de una distribución normal estándar.
def standard_normal_pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)

# Utiliza quad de SciPy para calcular numéricamente la integral desde menos infinito hasta zz de la función de densidad de probabilidad
def probit(z):
    result, _ = quad(standard_normal_pdf, -np.inf, z)
    return result

# Crear valores de x
x = np.linspace(-5, 5, 100)

# Calcular los valores de la función probit para los valores de x
y = [probit(xi) for xi in x]

# Dibujar la función probit
import matplotlib.pyplot as plt
plt.plot(x, y, label='Función Probit', color='purple')
plt.title('Función Sigmoide utilizada en Regresión Probit', fontsize=15)
plt.xlabel('z', fontsize=15)
plt.ylabel('Probit(z)', fontsize=15)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

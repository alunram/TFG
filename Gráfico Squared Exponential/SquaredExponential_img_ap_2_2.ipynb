```python
import numpy as np
import matplotlib.pyplot as plt

# Calcula la función de covarianza entre x e y, con parámetro length_scale
def squared_exponential_covariance(xp, xq, length_scale):
    sqdist = np.sum(xp**2, 1).reshape(-1, 1) + np.sum(xq**2, 1) - 2 * np.dot(xp, xq.T)

    return np.exp(-0.5 / length_scale**2 * sqdist)

# Define el espacio
x = np.linspace(-5, 5, 100).reshape(-1, 1)
length_scale = 1.2

# Calcula la matriz de covarianza
K = squared_exponential_covariance(x, x, length_scale)

# Dibuja el gráfico
plt.figure(figsize=(8, 6))
plt.imshow(K, extent=(-5, 5, -5, 5), origin='lower', cmap='viridis')
plt.colorbar(label='Covarianza')
plt.title('Función de Covarianza "Squared Exponential"')
plt.xlabel(r'$x_p$')  # LaTeX syntax
plt.ylabel(r'$x_q$')  # LaTeX syntax
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calcula la funci칩n de covarianza entre x e y, con par치metro length_scale
def squared_exponential_covariance(xp, xq):
    sqdist = np.sum(xp**2, 1).reshape(-1, 1) + np.sum(xq**2, 1) - 2 * np.dot(xp, xq.T)

    return np.exp(-0.5 * sqdist)

# Define el espacio
x = np.linspace(-5, 5, 100).reshape(-1, 1)

# Calcula la matriz de covarianza
K = squared_exponential_covariance(x, x)

# Dibuja el gr치fico
plt.figure(figsize=(8, 6))
plt.imshow(K, extent=(-5, 5, -5, 5), origin='lower', cmap='viridis')
plt.colorbar(label='Covarianza')
plt.title('Funci칩n de Covarianza "Squared Exponential"')
plt.xlabel(r'$x_p$')  # LaTeX syntax
plt.ylabel(r'$x_q$')  # LaTeX syntax
plt.show()
```
